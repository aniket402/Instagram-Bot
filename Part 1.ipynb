## Q5 Like/Unlike Posts

```python
#locating the text box , search for the handle  

search_box=driver.find_element_by_class_name('x3qfX')
search_box.clear()
search_box.send_keys('dilsefoodie')
```

```python
#clicking on the handle

g=driver.find_element_by_class_name('z556c')
g.click()
```

```python
# Liking the top 30 posts of dilsefoodie

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

# Clicking on the first post
d=driver.find_element_by_class_name('_9AhH0')
d.click()

# liking the first post
caption=driver.find_element_by_xpath('//span[contains(@class,"fr66n")]/button')
caption.click()

# Liking the remaining 29 posts
for i in range(29):
    
    # Moving to the next post by clicking on forward arrow
    wait = WebDriverWait(driver, 30)
    lap = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 
                                          'coreSpriteRightPaginationArrow')))
    lap.click()
    
    # Liking the post 
    wait = WebDriverWait(driver, 30)
    lap1 = wait.until(EC.presence_of_element_located((By.XPATH, 
                                              '//span[contains(@class,"fr66n")]/button')))
    lap1.click()
    
#closing the last post
a=driver.find_elements_by_tag_name('path')
a[24].click()
```

```python
# Unliking the top 30 posts of dilsefoodie

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

# Clicking on the first post
d=driver.find_element_by_class_name('_9AhH0')
d.click()

# Uniking the first post
caption=driver.find_element_by_xpath('//span[contains(@class,"fr66n")]/button')
caption.click()

# Liking the remaining 29 posts
for i in range(29):
    
    # Moving to the next post by clicking on forward arrow
    wait = WebDriverWait(driver, 30)
    lap = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 
                                          'coreSpriteRightPaginationArrow')))
    lap.click()
    
    # Uniking the post 
    wait = WebDriverWait(driver, 30)
    lap1 = wait.until(EC.presence_of_element_located((By.XPATH, 
                                              '//span[contains(@class,"fr66n")]/button')))
    lap1.click()
    
#closing the last post
a=driver.find_elements_by_tag_name('path')
a[24].click()
```

## Q6 A. Extract list of followers

```python
# Locating the text box and searching the required handle. 

import time
f=driver.find_element_by_class_name('x3qfX')
f.clear()
f.send_keys('foodtalkindia')
time.sleep(2) 

#click on the handle.
g=driver.find_element_by_class_name('z556c')
g.click()
```

```python
# Open the list of followers.

h=driver.find_element_by_partial_link_text('followers')
h.click()
```

```python
#Printing the first 500 followers 

handles=['foodtalkindia','sodelhi']

for handle in handles:
    url='https://www.instagram.com/'+handle+'/'
    driver.get(url)
    
    # Open the list of followers.

    followers_list=driver.find_element_by_partial_link_text('followers')
    followers_list.click()
    
    #locating the followers box
    f  = driver.find_element_by_xpath("//div[@class='isgrP']")
    followers_2 = []
    scroll = 0
    while scroll < 15:
        follower = driver.find_elements_by_xpath("//*[contains(@class, 'd7ByH')]")
        for i in follower:
            followers_2.append(i.text)
        driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;', f)
        time.sleep(2)
        scroll += 1
        
    print('NAME OF HANDLE',' : ',handle)
    for i in range(1,501):
        print(i,'.',followers_2[i])
    
    #closing the post
    c=driver.find_elements_by_class_name("wpO6b ")
    c[1].click()
```

### Q7. Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
### 1. If You have already seen the story.
### 2. Or The user has no story.
### 3. Or View the story if not yet seen.

```python
#Type the concerned handle name 

handle_name='coding.ninjas'

#First, we find the name of all handles which have a story and save those names in the list.
c=driver.find_elements_by_class_name("eebAO")
handles_having_stories=[]
for i in c:
    handles_having_stories.append(i.text)
    
#Check if your concerned handle has a story or not.

if handle_name in handles_having_stories:
    
    d=driver.find_elements_by_class_name("puCyQ")
    story_already_seen_handles=[]
    
    #If the handle has a story, check whether you have seen it or not. 
    for i in d:
        story_already_seen_handles.append(i.text)
    if handle_name in story_already_seen_handles:
        print('You have ALREADY SEEN the story' + ' of '+ handle_name)
    else:
        print('You have NOT SEEN the story of ' + handle_name)
        
else:
    print('Error: NO STORY of '+handle_name+ ' there')
```
